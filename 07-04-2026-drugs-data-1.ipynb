```python
from scipy.stats import shapiro

for column in data.columns:
    stat, p = shapiro(data[column])
    print(f"{column}: p-value = {p}")
```

```python
##Interpretation:
# If p > 0.05 → Normal (Parametric) (Drug-1 and Drug-2)
#If p < 0.05 → Not Normal (Non-parametric)
#If Parametric → use ANOVA
# If Non-parametric → use Kruskal-Wallis Test
```

```python
#if parametric
from scipy.stats import f_oneway

f_oneway(*[data[col] for col in data.columns])
```

```python
#if non-parametric
from scipy.stats import kruskal
kruskal(*[data[col] for col in data.columns])
```
